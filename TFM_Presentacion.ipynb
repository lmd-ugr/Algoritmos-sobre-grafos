## Building-up / Cutting-down

```python
g=Grafo([1,2,3,4])

for i in g.vertices:
    for j in g.vertices:
        if i > j:
            g.añadir_arista(i,j)
            
g.añadir_arista(1,5)
g.añadir_arista(2,5)

g.dibujar('circo')
```

```python
Constructivo(g, True)
```

```python
Constructivo(g).dibujar('circo')
```

```python
Destructivo(g, True)
```

```python
Destructivo(g).dibujar('circo')
```

## Kruskal

```python
g=Grafo([1,2,3,4])

for i in g.vertices:
    for j in g.vertices:
        if i > j:
            g.añadir_arista(i,j)
            
g.añadir_arista(1,5)
g.añadir_arista(2,5)


g.ponderado([1,4,7,1,8,9,2,8])
g.dibujar_ponderado('circo')
```

```python
Kruskal_Constructivo(g,True)
```

```python
Kruskal_Destructivo(g).dibujar_ponderado()
```

```python
Kruskal_Destructivo(g,True)
```

## Dijkstra

```python
g=Grafo([],[(1,2), (2,3), (3,4), (4,1)])
g.ponderado([2,1,3,1])
g.dibujar_ponderado('circo')
```

```python
Dijkstra(g,1)
```

```python
# Por simplicidad para el algoritmo hemos tomado el vector incial con todos sus elementos infinito salvo el del vértice inicial

Dijkstra(g,1,True)
```

Veamos qué ocurre si cambiamos el peso de la arista (3,4) por -3

```python
g=Grafo([],[(1,2), (2,3), (3,4), (4,1)])
g.ponderado([2,1,-3,1])
g.dibujar_ponderado('circo')
```

```python
Dijkstra(g,1)
```

Como el algoritmo de Dijkstra no detecta los ciclos negativos, hemos incluido otros dos: **Bellman-Ford** y **Floyd-Warshall**

```python
Bellman_Ford(g,1)
```

```python
Floyd_Warshall(g)
```
