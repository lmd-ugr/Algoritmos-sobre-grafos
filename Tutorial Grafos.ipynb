## Grafos Eulerianos

### Algoritmo de Fleury 

El algoritmo de Fleury se utiliza para calcular el camino o circuito de Euler, en caso de haberlo, en un grafo cualquiera. El funcionamiento de la función Fleury(g) que hemos implementado consiste en comprobar primero que tipo de grafo tenemos, si es Euleriano, si tiene un camino de Euler o ninguna de las anteriores. Para ello, utilizamos nuestro método 'g.es_euleriano()', que devuelve el valor 1, 2 ó -1 en caso de que el grafo estudiado tenga un circuito Euleriano, un camino de Euler o ninguno de los dos, respectivamente.

Así, el primer paso en Fleury(g), es conocer el vértice por el que empezaremos nuestro circuito o camino, que lo obtendremos en función del valor devuelto en 'g.es_euleriano()'. Si no existiese niguno de ellos, la función retornaría False.

```python
from Fleury import *

g=Grafo([1,2], [(1,2), (1,3), (2,3), (3,4), (3,5), (4,5)])
g.es_euleriano()
```

Para conocer si un grafo es o no Euleriano, hemos utilizado un método de nuestra clase grafos aún no mencionado: 'g.conexo()', que nos devuelve True o False según el número de componentes conexas que tenga el grafo, las cuáles se calculan con el método 'g.componentes_conexas()'.

```python
g.componentes_conexas()
```

```python
g.conexo()
```

Por tanto, hemos visto que el grafo g contiene un circuito de Euler. Veamos cómo funciona el algoritmo de Fleury en este caso.

```python
Fleury(g, True)
```

```python
Fleury(g)
```

### Algoritmo de Hierholzer -- Grafos Eulerianos

El algoritmo de Hierholzer es una alternativa al algoritmo de Fleury. La función 'Hierholzer(g)' solo requiere el grafo a estudiar como entrada, comprueba si el grafo contiene o no circuitos o caminos de Euler y utiliza el método 'g.ciclos()' para devolvernos la solución.

El método 'g.ciclos(long)' nos devuelve los ciclos de g de longitud (en vértices) estrictamente mayor que 'long'. El método 'g.ciclos()', a su vez, consiste en estudiar los caminos simples distintos desde un vértice a uno adyacente con él y unificarlos para formar un ciclo. Esto es posible gracias al método 'g.caminos_simples(u,v)', también implementado en nuestra clase 'grafos'. Este método recibe como entrada los vértices inicial y final y devuelve todos los caminos simples entre ellos.

```python
from Hierholzer import *

g=Grafo([], [(1,2), (1,3), (1,4), (1,5), (2,3), (4,5), (5,6), (5,7)])
g.dibujar('neato')
```

```python
Hierholzer(g)
```

```python
Hierholzer(g,True)
```

## Caminos de peso mínimo

### Algoritmo de Dijkstra

Este algoritmo se utiliza para encontrar el camino de menor peso entre un vértice tomado como inicial y el resto. Para ello trabajamos con el vector de distancias, donde cada elemento de dicho vector indica la distancia entre el vértice al que representa la posición que ocupa y el vértice tomado como nodo inicial. En este caso, entendemos la distancia entre dos vértices como el peso del camino que los conecta.

La función recibe como argumentos el grafo a estudiar y el vértice inicial y devuelve dos vectores:
1. El vector de distancias mínimas.
2. El vector de ruta que nos indica cómo llegar desde el vértice inicial al resto.

```python
from Dijkstra import *

g=Grafo([],[(1,2),(2,3),(3,4),(1,5),(2,5),(3,5),(3,6),(4,6),(5,6)])
g.ponderado([4,2,8,2,1,7,9,9,3])
g.dibujar_ponderado('circo')
```

```python
Dijkstra(g,1,True)
```

```python
for i in g.vertices:
    print('Vértice ', str(i), ': ')
    H=Dijkstra(g,i)
    print(H, '\n')
```

### Algoritmo de Bellman-Ford

Este algoritmo es una alternativa al algoritmo de Dijkstra, ya que éste último no nos informa de la aparición de ciclos negativos y, por tanto, puede devolver resultados erróneos. Al igual que en Dijkstra, trabajamos con el vector de distancias, ingresamos como argumentos de la función, el grafo y el vértice inicial y obtendremos el vector de distancias mínimas, el vector de ruta y, en caso de haberlos, el mensaje 'Hay ciclo neagtivo'.

```python
from Bellman_Ford import *

g=Grafo([], [(1,2), (1,3), (2,3), (2,4), (3,4)])
g.ponderado([5,10,-5,10,10])
g.dibujar_ponderado('circo')
```

```python
Bellman_Ford(g,1)
```

```python
Bellman_Ford(g,1,True)
```

### Algoritmo de Floyd-Warshall

Al igual que Bellman-Ford, este algoritmo es una alternativa a Dijkstra para poder detectar ciclos negativos. En este caso trabajamos con una matriz distancias D donde cada elemento D_ij_ es el peso del camino mínimo actual entre el vértice i y eñ vértice j. Las distancias se irán actualizando recalculando el peso de los caminos entre i y j tomando cada vez un vértice distinto como nodo intermedio.

En este caso, solo requerimos el grafo como argumento de nuestra función y nos devolverá la matriz de distancias mínimas (en caso de no haber ciclo negativo) y la matriz de ruta.

```python
from Floyd_Warshall import *

g=Grafo([], [(1,2), (1,3), (2,3)])
g.ponderado([1,2,3])
g.dibujar_ponderado('circo')
```

```python
Floyd_Warshall(g)
```

```python
g=Grafo([], [(1,2), (1,3), (2,3)])
g.ponderado([1,2,-3])
g.dibujar_ponderado('circo')
```

```python
Floyd_Warshall(g, True)
```

## Optimización de flujos

### Algoritmo de Ford-Fulkerson

Este algoritmo de aplica sobre grafos ponderados donde los pesos no son números, sino listas de dos elementos. El primer elemento de la lista asociada a la arista (i,j) es el peso asociado en la dirección i-->j, mientras que el segundo peso es el que debemos tener en cuenta para recorrer dicha arista en la dirección j-->i.

La función recibe como argumentos el grafo estudiado y los vértices inicial y final (fuente y sumidero, respectivamente) y nos devuelve un grafo donde los pesos de las aristas (en este caso ya no son listas, sino valores) indican el flujo óptimo de producto que puede circular por dicha arista.

```python
from Ford_Fulkerson import *

g=Grafo([], [(1,2), (1,3), (1,4), (2,3), (3,4), (3,5), (4,5)])
g.ponderado([[10,10], [10,20], [10,0],[30,10], [10,15], [20,20],[10,10]])
g.dibujar_ponderado()
```

```python
Ford_Fulkerson(g,1,5).dibujar_ponderado()
```

```python
Ford_Fulkerson(g,1,5,True)
```
